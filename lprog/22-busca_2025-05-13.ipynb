```python
achar = 152
lista = sorted([ 103, 123, 152, 158, 173, 52, 73, 95, 98, 99])
lista
```

```python
i = 5
lista[i]
```

## Análise assintótica

- Método utilizado para avaliar o desempenho de um algoritmo em termos de tempo de execução e espaço de memória, especialmente para grandes volumes de dados
- O principal objetivo é encontrar o comportamento do algoritmo (a função matemática) em relação ao tempo de execução para o pior caso, o que chamamos de complexidade assintótica
- A análise assintótica se concentra em como o tempo de execução de um algoritmo cresce à medida que o tamanho da entrada (denotado por 'n') aumenta significativamente

- A complexidade de um algoritmo é expressa por uma função matemática f(n)
- A função é simplificada, considerando apenas o termo dominante, que é o termo que cresce mais rapidamente com o aumento de 'n'
- A complexidade é expressa utilizando a notação Big-Oh (Grande-O), denotada por O(f(n))

### Exemplo 1 : busca linear

- A função que descreve o tempo de execução é f(n) = 1 * n
- No pior caso, o algoritmo pode precisar percorrer todos os 'n' elementos da lista para encontrar o valor desejado
- O "1" representa o custo de cada operação individual na busca linear. Neste caso temos 1 operação de comparação para cada elemento
- Ao aplicar a análise assintótica, ignoramos a constante (1) e concluímos que a complexidade do algoritmo de busca linear é O(n), que representa uma complexidade linear

### Exemplo 2 : função de tempo hipotética

f(n) = an² + bn + c

- identifica-se o termo dominante n² , e o resto é ignorado
- resulta em uma complexidade de O(n²)
